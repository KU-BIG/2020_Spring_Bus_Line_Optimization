#출근시간 인기 정류장 거리

```r
morning <- rbind(on79, off79)
length(unique(morning$ID)) == length(morning$ID)
```

```r
#install.packages("geosphere")
#install.packages("tidyverse")
#install.packages("rlang")
devtools::install_github("r-lib/rlang", build_vignettes = TRUE)
```

```r
library(geosphere)
```

```r
colnames(morning) <- c("X", "ID", "Station_Name", "lat", "long", "number")
morning <- morning[order(morning$ID),]
morning <- morning[!duplicated(morning$ID),]
head(morning)
```

```r
morning_dist <- matrix(, nrow=length(morning$ID), ncol=length(morning$ID))
dimnames(morning_dist) <- list(morning$ID, morning$ID)
```

```r
for (i in 1:length(morning$ID)){
    for (j in 1:length(morning$ID)){
        x_lat <- morning$lat[i]
        x_long <- morning$long[i]
        y_lat <- morning$lat[j]
        y_long <- morning$long[j]
        morning_dist[i,j] <- distm(c(x_long, x_lat), c(y_long, y_lat), fun = distHaversine)
    }
}
```
